# 🧪 Age Validation in Cancer Mortality Data

## 🔍 Objective
Check for:
- Negative ages  
- Ages above 100 (approximate threshold)

---

## 📊 Reference Data (Google Estimates)

### Median Age of Cancer Patients at Death
- **Men:** ~200,100 deaths annually (ages 30–69)  
- **Women:** ~195,300 deaths annually (ages 30–69)  

---

## 📌 Insight
- Most fatal cases occur in **middle-aged adults (30–69 years)**  
- Highlights the **loss of productive years** due to cancer mortality  

---

## ✅ Validation Rule
- Age must be **≥ 0**  
- Age must be **≤ 100**  


```python
# Survival Validation
df["Survival_Months"].describe()
```

```python
df[df["Survival_Months"] < 0]
```

```python
# feature Engineering 

df["Diagnosis_Year"] = df["Diagnosis_Date"].dt.year
```

```python
df.head(5)
```

```python
# Creating Age Group
bins = [0,18,35,50,65,100]

labels = [
    "Child",
    "Young Adult",
    "Adult",
    "Senior",
    "Elder"
]



df["Age_group"] = pd.cut(
    df["Age"],
    bins = bins,
    labels = labels
    
)
```

```python
df["Survival_Category"] = pd.cut(
    df["Survival_Months"],
    bins=[0,12,24,36,100],
    labels=[
        "Low",
        "Medium",
        "High",
        "Very High"
    ]
    
)
```
